```python
df['reviews per month'] = df['reviews per month'].fillna(df['reviews per month'].mean())
```

```python
findNull(df)
```

```python
df['review rate number'] = df['review rate number'].fillna(df['review rate number'].mean())
```

```python
findNull(df)
```

```python
df['calculated host listings count'] = df['calculated host listings count'].fillna(1)
findNull(df)
```

```python
df['host_identity_verified'] = df['host_identity_verified'].fillna("NO IDEA")
findNull(df)
```

```python
verified_count = df['host_identity_verified'].value_counts()
verified_count
v = 0
no = 0
u = 0
for d in df['host_identity_verified']:
    if d == "NO IDEA":
        no+=1
    elif d == "verified":
        v+=1
    else:
        u+=1
        
print("===============================================")
print(f'verified: {v}\nunverified: {u},\nNO IDEA: {no}')
print("===============================================")

plt.pie(verified_count, labels=verified_count.index, autopct='%1.1f%%', colors=['cyan','pink','red'])
plt.tight_layout()
plt.legend(verified_count.index, loc='center left', bbox_to_anchor=(1.2,0.5), title="Host Identiy Verification")
plt.title("Distribution of Host's identiy verification")
plt.axis('equal')
plt.show()
```

```python
df = df.dropna(subset=['price','minimum nights','number of reviews'])
findNull(df)
```

```python
df['service fee'] = df['service fee'].replace('[$,]', '', regex=True)
df['service fee'] = pd.to_numeric(df['service fee'], errors='coerce')
df['service fee'] = df['service fee'].fillna(df['service fee'].mean())
```

<h3>HOST IDENTITY</h3>

```python
host_count = df['host_identity_verified'].value_counts()
plt.pie(host_count, labels=host_count.index, autopct='%1.1f%%', colors=['cyan','skyblue','red'])
plt.legend(host_count.index, title="Identity Status", loc='center left', bbox_to_anchor=(1,1))
plt.title("Host identity verification")
plt.show()
```

<h3>NEIGHBOORHOOD GROUP</h3>

```python
hood_count = df['neighbourhood group'].value_counts()
print(hood_count)

explode = [0.05] * len(hood_count)

plt.figure(figsize=(16,6))

# Pie Chart
plt.subplot(1, 2, 1)
plt.pie(
    hood_count,
    labels=hood_count.index,
    autopct='%1.1f%%',
    explode=explode,
    startangle=140
)
plt.legend(hood_count.index, loc='center left', bbox_to_anchor=(1, 0.5), title='Hood name')
plt.title("Neighbourhood Group Distribution - Pie Chart")
plt.axis('equal')

# Bar Chart
plt.subplot(1, 2, 2)
plt.bar(hood_count.index, hood_count.values, color='skyblue')
plt.title("Neighbourhood Group Distribution - Bar Chart")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

<h3>Bookable</h3>

```python
book_count = df['instant_bookable'].value_counts()
book_count.index = book_count.index.map({False: 'No', True: 'Yes'})
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.pie(book_count, labels=book_count.index, autopct='%1.1f%%')
plt.legend(book_count.index, title='Legend', bbox_to_anchor=(1,1), loc='center left')
plt.title('Instantly Bookable Chances (Pie)')

plt.subplot(1,2,2)
plt.bar(book_count.index, book_count.values)
plt.title('Instantly Bookable Chances (Bar)')
plt.show()
```
