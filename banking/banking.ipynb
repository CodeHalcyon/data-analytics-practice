```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Banking.csv")
df.head(5)
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Get the value counts of Nationality
nationality_counts = df['Nationality'].value_counts().reset_index()

# Rename columns for easier reference
nationality_counts.columns = ['Nationality', 'Count']

# Create a line plot
sns.barplot(data=nationality_counts, x='Nationality', y='Count')

# Show the plot
plt.xticks(rotation=90)  # Rotate x-axis labels if needed for readability
plt.show()
```

```python
loyalty_count =  df['Loyalty Classification'].value_counts()

plt.pie(loyalty_count, labels=loyalty_count.index, autopct='%1.1f%%',colors=['cyan','skyblue','pink','gray'])
plt.show()
```
