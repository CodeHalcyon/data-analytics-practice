```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("movies.csv")
df.drop_duplicates()
```

```python
df['ONE-LINE'] = df['ONE-LINE'].str.replace("\n","")
df['GENRE'] = df['GENRE'].str.replace("\n","")
df['STARS'] = df['STARS'].str.replace("|","")
df['STARS'] = df['STARS'].str.replace("\n","")

# df.fillna()

df['RATING'] = df['RATING'].fillna("")
df['VOTES'] = df['VOTES'].fillna("")
# df.isnull().sum()
df
```

```python
df = df.drop(columns=["Gross","RunTime"])
```

```python
df.isnull().sum()
```

```python
df[["Director/Actor","Actor2","Actor3","Actor4"]] = df['STARS'].str.split(",",n=3, expand=True)
df.Actor2 = df.Actor2.fillna("")
df.Actor3 = df.Actor3.fillna("")
df.Actor4 = df.Actor4.fillna("")
df
```
